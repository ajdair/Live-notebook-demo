## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sample points
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.1, 100)
```

```python
# Plot the pints
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sample Data', color='blue', linestyle='-')
plt.title('Sample Data Visualization')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
